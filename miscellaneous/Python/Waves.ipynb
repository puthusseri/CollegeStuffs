```python
import numpy as np
import sys
t = np.linspace(-np.pi, np.pi, 100)
k = np.arange(1, 99.0)
f = np.zeros_like(t)
```

```python
for i in range(len(t)):
   f[i] = np.sum(np.sin(2 * np.pi * k * t[i])/k)
f = (-2 / np.pi) * f
```

# Plotting both the triangle wave and the swatooth wave

```python
import matplotlib as plt
from  matplotlib.pyplot import *
plot(t, f, lw=1.0)
plot(t, np.abs(f), lw=2.0)
show()
```
